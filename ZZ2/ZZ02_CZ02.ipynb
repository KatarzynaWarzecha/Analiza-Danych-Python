### Zadnie 2: Prezentacja danych w formie tabelarycznej [10 pkt]

Przygotuj trzy tablice szeregów rodzielczych przedziałowych częstości wskaźnika GDP growth (annual %) dla poszczególnych lat.

 - 2 poziomu: Year (wszystkie lata dla każdego z rejonów)

 - Kolejne przedziały (-20, -15]	(-15, -10] ... (30, 35] zdefiniowane jako IntervalIndex

Trzecia tablica (`recession_table_3`) powinna zawierać szeregi skonstruowane indywidualnie dla każdego stopnia zamożności.

```python
# Całość danych
# YOUR CODE HERE
#raise NotImplementedError()
pd.options.mode.chained_assignment = None  

out = pd.cut(recession_data_1['2006'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True)
recession_table_1 = pd.DataFrame(out.value_counts()).T
recession_table_1 = recession_table_1.reindex(sorted(recession_table_1.columns), axis=1)
recession_table_1.loc['2007'] = pd.cut(recession_data_1['2007'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True).value_counts()
recession_table_1.loc['2008'] = pd.cut(recession_data_1['2008'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True).value_counts()
recession_table_1.loc['2009'] = pd.cut(recession_data_1['2009'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True).value_counts()
recession_table_1.loc['2010'] = pd.cut(recession_data_1['2010'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True).value_counts()
recession_table_1.loc['2011'] = pd.cut(recession_data_1['2011'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True).value_counts()

recession_table_1.loc['2006'] = recession_table_1.loc['2006'] /recession_table_1.loc['2006'].sum()
recession_table_1.loc['2007'] = recession_table_1.loc['2007'] /recession_table_1.loc['2007'].sum()
recession_table_1.loc['2008'] = recession_table_1.loc['2008'] /recession_table_1.loc['2008'].sum()
recession_table_1.loc['2009'] = recession_table_1.loc['2009'] /recession_table_1.loc['2009'].sum()
recession_table_1.loc['2010'] = recession_table_1.loc['2010'] /recession_table_1.loc['2010'].sum()
recession_table_1.loc['2011'] = recession_table_1.loc['2011'] /recession_table_1.loc['2011'].sum()
recession_table_1

display_recession_table_1=recession_table_1[:].round(decimals=2)
display_recession_table_1
```

```python
# Podział na rejony
# YOUR CODE HERE
#raise NotImplementedError()
pd.options.mode.chained_assignment = None  
recession_table_2 = pd.DataFrame()
regions = ['Asia', 'Europe', 'Africa', 'Oceania', 'Americas']
for x in regions:
    region = recession_data_1[recession_data_1['Region Name'] == x]

    out = pd.cut(region['2006'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True)
    recession_table = pd.DataFrame(out.value_counts()).T
    recession_table = recession_table.reindex(sorted(recession_table.columns), axis=1)
    recession_table.loc['2007'] = pd.cut(recession_data_1['2007'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True).value_counts()
    recession_table.loc['2008'] = pd.cut(recession_data_1['2008'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True).value_counts()
    recession_table.loc['2009'] = pd.cut(recession_data_1['2009'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True).value_counts()
    recession_table.loc['2010'] = pd.cut(recession_data_1['2010'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True).value_counts()
    recession_table.loc['2011'] = pd.cut(recession_data_1['2011'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True).value_counts()

    recession_table.loc['2006'] = recession_table.loc['2006'] /recession_table.loc['2006'].sum()
    recession_table.loc['2007'] = recession_table.loc['2007'] /recession_table.loc['2007'].sum()
    recession_table.loc['2008'] = recession_table.loc['2008'] /recession_table.loc['2008'].sum()
    recession_table.loc['2009'] = recession_table.loc['2009'] /recession_table.loc['2009'].sum()
    recession_table.loc['2010'] = recession_table.loc['2010'] /recession_table.loc['2010'].sum()
    recession_table.loc['2011'] = recession_table.loc['2011'] /recession_table.loc['2011'].sum()
    recession_table_2 = pd.concat([recession_table_2,recession_table])

asia = pd.concat([recession_table_2.iloc[[0,1,2,3,4,5]]],keys=["Asia"])
europe = pd.concat([recession_table_2.iloc[[6,7,8,9,10,11]]],keys=["Europe"])
africa = pd.concat([recession_table_2.iloc[[12,13,14,15,16,17]]],keys=["Africa"])
oceania = pd.concat([recession_table_2.iloc[[18,19,20,21,22,23]]],keys=["Oceania"])
americas = pd.concat([recession_table_2.iloc[[24,25,26,27,28,29]]],keys=["Americas"])
recession_table_2 = pd.concat([asia,europe,africa,oceania,americas])

display_recession_table_2=recession_table_2[:].round(decimals=2)
display_recession_table_2
```

```python
# Podział pod względem stopnia zamożności
# YOUR CODE HERE
#raise NotImplementedError()
pd.options.mode.chained_assignment = None  
recession_table_3 = pd.DataFrame()
group = ['Low', 'Lower middle', 'Upper middle', 'High']
for x in group:
    group = recession_data_1[recession_data_1['Income Group'] == x]

    out = pd.cut(group['2006'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True)
    recession_table = pd.DataFrame(out.value_counts()).T
    recession_table = recession_table.reindex(sorted(recession_table.columns), axis=1)
    recession_table.loc['2007'] = pd.cut(recession_data_1['2007'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True).value_counts()
    recession_table.loc['2008'] = pd.cut(recession_data_1['2008'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True).value_counts()
    recession_table.loc['2009'] = pd.cut(recession_data_1['2009'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True).value_counts()
    recession_table.loc['2010'] = pd.cut(recession_data_1['2010'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True).value_counts()
    recession_table.loc['2011'] = pd.cut(recession_data_1['2011'],bins=[-20,-15,-10,-5,0,5,10,15,20,25,30,35],right=True).value_counts()

    recession_table.loc['2006'] = recession_table.loc['2006'] /recession_table.loc['2006'].sum()
    recession_table.loc['2007'] = recession_table.loc['2007'] /recession_table.loc['2007'].sum()
    recession_table.loc['2008'] = recession_table.loc['2008'] /recession_table.loc['2008'].sum()
    recession_table.loc['2009'] = recession_table.loc['2009'] /recession_table.loc['2009'].sum()
    recession_table.loc['2010'] = recession_table.loc['2010'] /recession_table.loc['2010'].sum()
    recession_table.loc['2011'] = recession_table.loc['2011'] /recession_table.loc['2011'].sum()
    recession_table_3 = pd.concat([recession_table_3,recession_table])

low = pd.concat([recession_table_3.iloc[[0,1,2,3,4,5]]],keys=["Low"])
lower_middle = pd.concat([recession_table_3.iloc[[6,7,8,9,10,11]]],keys=["Lower middle"])
upper_middle = pd.concat([recession_table_3.iloc[[12,13,14,15,16,17]]],keys=["Upper middle"])
high = pd.concat([recession_table_3.iloc[[18,19,20,21,22,23]]],keys=["High"])
recession_table_3 = pd.concat([low,lower_middle,upper_middle,high])

display_recession_table_3=recession_table_3[:].round(decimals=2)
display_recession_table_3
```

#### c) Dyskusja wyników

Zastanów się jakie są mocne i słabe strony poszczególnych typów wykresów wykonanych w tym zadaniu. W jakich przypadkach wykorzystanie konkretnych typów wykresów do wizualizacji może sprawiać problemy?

- Mocną stroną histogramów oraz wykresów typu boxplot jest ich znacząca wartość informacyjna oraz zwiększona względem innych typów łatowść w precyzyjnej interpretacji otrzymanych wyników. W mojej opinii, te dwa wspomniane typy są, w przypadku tego zadania, najbardziej użyteczne.
- Zaletą wykresów typu boxplot i violinplot jest możliwość pozyskania informacji statystycznych o zbiorach takich, jak m.inn wartość średnia.
- Wadą wykresów typu stripplot i swarmplot jest duże nagromadzenie punktów w jednym, niezbyt dużym obszarze. Utrudnia to pełną interpretację otrzymanych rezultatów i w przypadku większych zbiorów danych może skutkować zaciemnieniem analizowanych danych i trudnościami z poprawnym odczytaniem wyników. Wykresy tego typu lepiej sprawdzą się dla mniejszych zbiorów, takich o niewielkiej liczbie etykiet.
- W przypadku wykresów typu histplot, podobnie jak dla swarmplot i stripplot, problem mógłby stanowić zbiór o dużej kardynalności - tu również mogłoby to skutkować trudnością w poprawnym przeprowadzeniu analizy.
