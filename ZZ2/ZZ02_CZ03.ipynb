#### a) Obliczanie wartości wybranych opisowych charakterystyk rozkładu

Korzystając z danych zawartych w tablicy `recession_table_1` oblicz dla każdego roku następujące opisowe charakterystyki rozkładu:
- wartość średnia (ang. mean),
- odchylenie standardowe (ang. standart deviation),
- kwartyle (ang. quartiles),
- moda (ang. mode).

Zestaw wyniki w tablicy `descriptive_parameters_2`.

Indeks:
 - zawiera następujące nazwy wierszy: mean, std, Q1, Q2, Q3, mode.
 
Kolumny:
 - lata (osobna kolumna dla każdego roku opatrzona nagłówkiem 2006, 2007 itd. sformatowanym jako str)
 
Uwaga! W celu otrzymania maksymalnej liczby punktów otrzymane wyniki powinny być w całości wynikiem pracy napisanego kodu, bez manualnego określania, w którym przedziale znajdują się określone statystyki itp.

```python
# YOUR CODE HERE
#raise NotImplementedError()
# https://www.statystyka-zadania.pl/szereg-przedzialowy/

year = ['2006','2007','2008','2009','2010','2011']
mean=[]
std = []
mode = []
q1 = []
cumSum = pd.DataFrame()
q2 = []
q3 = []


for x in year:
    #MEAN
    srednia = sum(recession_table_1.index.mid*recession_table_1[x])/recession_table_1[x].sum()
    mean.append(srednia)
    
    #STD
    war = sum( (recession_table_1.index.mid-srednia)*(recession_table_1.index.mid-srednia)*recession_table_1[x]) /recession_table_1[x].sum()
    std.append(math.sqrt(war))
    
    #MODE
    dom = recession_table_1.index[recession_table_1[x]==max(recession_table_1[x])] #szukam indeksu dla maksymalnej wartości
    xD = dom.left
    hD = dom.length
    nD = max(recession_table_1[x])
    rows = recession_table_1.loc[dom]
    a = recession_table_1.shift(1).loc[dom]
    b =recession_table_1.shift(-1).loc[dom]
    nDbefore = a[x]
    nDafter  = b[x]
    mode.append((xD + (nD - nDbefore)/((nD-nDafter)+(nD - nDbefore)) * hD).item())  #bierzemy tylko wartości, bez indeksu
    
    #q1 q2 q3
    #cumSum[x] = recession_table_1[x]
    cumSum[x] = recession_table_1[x].cumsum()  #liczebność skumulowana
    nQ1 = recession_table_1[x].sum() / 4  #49,5
    nQ2 = recession_table_1[x].sum() / 2  #99
    nQ3  = 3* recession_table_1[x].sum()/4   #148,5
    przedzial_nQ1 = cumSum.loc[cumSum[x]>nQ1].head(1)   #123
    przedzial_nQ2 = cumSum.loc[cumSum[x]>nQ2].head(1)   #123
    przedzial_nQ3 = cumSum.loc[cumSum[x]>nQ3].head(1)   #187

    before1 =  cumSum[x].shift(1).loc[przedzial_nQ1.index]
    before2 =  cumSum[x].shift(1).loc[przedzial_nQ2.index]
    before3 =  cumSum[x].shift(1).loc[przedzial_nQ3.index]

    q1.append((przedzial_nQ1.index.left + ( nQ1  - before1   )/recession_table_1[x].loc[przedzial_nQ1.index]   * przedzial_nQ1.index.length).item())
    q2.append((przedzial_nQ2.index.left + ( nQ2  - before2  )/recession_table_1[x].loc[przedzial_nQ2.index]   * przedzial_nQ2.index.length).item())
    q3.append((przedzial_nQ3.index.left + ( nQ3  - before3  )/recession_table_1[x].loc[przedzial_nQ3.index]   * przedzial_nQ3.index.length).item())



d = {'Index': ['mean', 'std','Q1','Q2','Q3','mode'],
     '2006':[mean[0],std[0],q1[0],q2[0],q3[0],mode[0]],
     '2007':[mean[1],std[1],q1[1],q2[1],q3[1],mode[1]],
     '2008':[mean[2],std[2],q1[2],q2[2],q3[2],mode[2]],
     '2009':[mean[3],std[3],q1[3],q2[3],q3[3],mode[3]],
     '2010':[mean[4],std[4],q1[4],q2[4],q3[4],mode[4]],
     '2011':[mean[5],std[5],q1[5],q2[5],q3[5],mode[5]]
    }
descriptive_parameters_2 = pd.DataFrame(d).set_index('Index').rename_axis('', axis=1)
descriptive_parameters_2


cumSum['2006_rf'] = recession_table_1['2006']/sum(recession_table_1['2006'])
cumSum['2006_crf'] = cumSum['2006_rf'].cumsum()

cumSum['2007_rf'] = recession_table_1['2007']/sum(recession_table_1['2007'])
cumSum['2007_crf'] = cumSum['2007_rf'].cumsum()

cumSum['2008_rf'] = recession_table_1['2008']/sum(recession_table_1['2008'])
cumSum['2008_crf'] = cumSum['2008_rf'].cumsum()

cumSum['2009_rf'] = recession_table_1['2009']/sum(recession_table_1['2009'])
cumSum['2009_crf'] = cumSum['2009_rf'].cumsum()

cumSum['2010_rf'] = recession_table_1['2010']/sum(recession_table_1['2010'])
cumSum['2010_crf'] = cumSum['2010_rf'].cumsum()

cumSum['2011_rf'] = recession_table_1['2011']/sum(recession_table_1['2011'])
cumSum['2011_crf'] = cumSum['2011_rf'].cumsum()

descriptive_parameters_2
```
