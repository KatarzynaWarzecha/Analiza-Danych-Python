```python
# Miejsce na dodatkowy kod
recession_data_1 = pd.read_csv('recession_data_1.csv')
recession_data_1 = recession_data_1.set_index('Country Code')
recession_data_1
kolory = ['gold', 'cornflowerblue','grey','palegreen','lightcoral']
y_lim = [-30,40]
tekstowe = ['Country Name', 'Region Name', 'Income Group']
numeryczne =['2006', '2007', '2008', '2009', '2010', '2011']
fig, axes = plt.subplots(1,1,figsize = (7,4), facecolor='white')

plot_data = pd.melt(recession_data_1.drop(['Region Name','Income Group'],axis=1),var_name='Year', value_name='GDP annual', value_vars=['2006', '2007', '2008', '2009', '2010', '2011'])
meanprops = {"marker":"o","markerfacecolor":"grey", "markeredgecolor":"black", "markersize":"5"}
#plot_data
x=sns.boxplot(data=plot_data, x= 'Year', y='GDP annual', color="gold", ax = axes, showmeans=True, meanprops=meanprops, zorder=2)
axes.set_ylim(y_lim)
axes.grid(axis='y',zorder=0)
axes.axhline(0,c='red', zorder=1)
axes.set_title("Całość Danych")
fig.savefig('boxplot.png')

fig2, axes2 = plt.subplots(1,1,figsize = (7,4), facecolor='white')
plot_data = pd.melt(recession_data_1,id_vars = tekstowe, value_vars= numeryczne, var_name='Year', value_name = 'GDP annual')
sns.boxplot(data=plot_data, x= 'Year', y='GDP annual', hue='Region Name',palette = kolory, ax = axes2, showmeans=True, meanprops=meanprops, zorder=2)
axes2.set_ylim(y_lim)
axes2.grid(axis='y',zorder=0)
axes2.axhline(0,c='red', zorder=1)
axes2.set_title("Regiony")
axes2.set_axisbelow(True)
sns.move_legend(axes2,loc='upper right', ncol=5,frameon=True,columnspacing=1)
fig2.savefig('boxplot_region.png')
new = recession_data_1.loc[recession_data_1['Region Name']=='Europe',:]

fig3, axes3 = plt.subplots(1,1,figsize = (7,4), facecolor='white')
plot_data = pd.melt(recession_data_1,id_vars = tekstowe, value_vars= numeryczne, var_name='Year', value_name = 'GDP annual')
sns.boxplot(data=plot_data, x= 'Year', y='GDP annual', hue='Income Group',palette = 'coolwarm', ax = axes3, showmeans=True, meanprops=meanprops, zorder=2)
axes3.set_ylim(y_lim)
axes3.grid(axis='y',zorder=0)
axes3.axhline(0,c='red', zorder=1)
axes3.set_title("Stopień Zamożności")
axes3.set_axisbelow(True)
sns.move_legend(axes3,loc='upper right', ncol=5,frameon=True,columnspacing=1)
fig3.savefig('boxplot_income.png')

fig4, ax = plt.subplots(1,1,figsize=(7,4), facecolor="white")
df = pd.melt(recession_data_1, id_vars = ["Income Group"], value_vars = recession_data_1.columns[3:])
ax.set_facecolor("white")
ax.set_ylim([-25,40])
ax.set_title("Income Groups")
ax.grid(True)
ax.axhline(y=0, color="red")
sns.stripplot(data = df, x="variable", y="value", ax=ax, hue="Income Group", dodge = True, palette = "coolwarm")
ax.legend(title="Income Group", ncols=4)
ax.set_ylabel("GPD growth (annual %)")
ax.set_xlabel("Year")
fig4.savefig('swarm_plot_income.png')
```
